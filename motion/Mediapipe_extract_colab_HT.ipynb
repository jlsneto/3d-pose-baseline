```
import cv2
import mediapipe as mp
import cereja as cj
from copy import deepcopy

mp_drawing = mp.solutions.drawing_utils
mp_holistic = mp.solutions.holistic

# For video input:
def extract_keypoints(video):
    cap = cv2.VideoCapture(video)
    pose_data = []
    left_hand_data = []
    right_hand_data = []
    face_data = []

    with mp_holistic.Holistic(min_detection_confidence=0.5,
        min_tracking_confidence=0.5,) as holistic:
        while cap.isOpened():
            success, image = cap.read()
            if not success:
                print("Ignoring empty camera frame.")
                # If loading a video, use 'break' instead of 'continue'.
                break

            # Flip the image horizontally for a later selfie-view display, and convert
            # the BGR image to RGB.
            image = cv2.cvtColor(cv2.flip(image, 1), cv2.COLOR_BGR2RGB)
            # To improve performance, optionally mark the image as not writeable to
            # pass by reference.
            image.flags.writeable = False
            results = holistic.process(image)
            
            
            if results.pose_landmarks:
                pose_data.append([(i.x, i.y, i.z) for i in results.pose_landmarks.landmark])
            else:
                pose_data.append([[0.0,0.0,0.0]] * 25)

            if results.left_hand_landmarks:
                left_hand_data.append([(i.x, i.y, i.z) for i in results.left_hand_landmarks.landmark])
            else:
                left_hand_data.append([[0.0,0.0,0.0]] * 21)
            
            if results.right_hand_landmarks:
                right_hand_data.append([(i.x, i.y, i.z) for i in results.right_hand_landmarks.landmark])
            else:
                right_hand_data.append([[0.0,0.0,0.0]] * 21)

            if results.face_landmarks:
                face_data.append([(i.x, i.y, i.z) for i in results.face_landmarks.landmark])
            else:
                face_data.append([[0.0,0.0,0.0]] * 468)
            

        cap.release()
    
    return pose_data,left_hand_data,right_hand_data,face_data

def next_valid_coord(data,frame):
    for n_frame, frame_data in enumerate(data):
        if n_frame<frame:
            continue
        sum_frame = 0
        for n, coord in enumerate(frame_data):
            sum_frame += sum(coord)
        if sum_frame != 0:
            return n_frame

def last_valid_coord(data,frame):
    last_coord = 0
    for n_frame, frame_data in enumerate(data):
        if n_frame > frame:
            break
        sum_frame = 0
        for n, coord in enumerate(frame_data):
            sum_frame += sum(coord)
        if sum_frame != 0:
            last_coord = n_frame
    return last_coord


def interpolate_keypoints(list_data):
    
    pose_keypoints_data, hand_left_keypoints_data, hand_right_keypoints_data = [], [], []
    keypoints_list = []
    for i in range(len(list_data)):
        data = list_data[i]
        keypoints_data = []
        for n_frame, frame_data in enumerate(list_data[i]):
            part = []

            sum_coord = 0
            for n, coord in enumerate(frame_data):
                x, y, z = coord
                part.append((x, y, z))
                sum_coord += sum(coord)

            if sum_coord == 0:
                part = []
                last_valid = last_valid_coord(list_data[i],n_frame)
                next_valid = next_valid_coord(list_data[i],n_frame)
                if next_valid and last_valid:
                    last_contrib = (n_frame-last_valid)/(next_valid-last_valid)
                    next_contrib = (next_valid - n_frame)/(next_valid-last_valid)
                    
                    for n, coord in enumerate(frame_data):
                        x = list_data[i][last_valid][n][0]*next_contrib + list_data[i][next_valid][n][0]*last_contrib 
                        y = list_data[i][last_valid][n][1]*next_contrib + list_data[i][next_valid][n][1]*last_contrib 
                        z = list_data[i][last_valid][n][2]*next_contrib + list_data[i][next_valid][n][2]*last_contrib 
                        part.append((x, y, z))
                elif next_valid:
                    for n, coord in enumerate(frame_data):
                        x = list_data[i][next_valid][n][0]
                        y = list_data[i][next_valid][n][1]
                        z = list_data[i][next_valid][n][2]
                        part.append((x, y, z))
                else:
                    for n, coord in enumerate(frame_data):
                        x = list_data[i][last_valid][n][0]
                        y = list_data[i][last_valid][n][1]
                        z = list_data[i][last_valid][n][2]
                        part.append((x, y, z))

            keypoints_data.append(part)
        keypoints_list.append(keypoints_data)
    
    return keypoints_list


def put_hand_in_body(pose_data,hand_left_data,hand_right_data):
    
    for frame in range(len(hand_left_data)):
        part_r = []
        part_l = []
        l_disp = [hand_left_data[frame][0][0] - pose_data[frame][15][0],
                   hand_left_data[frame][0][1] - pose_data[frame][15][1],
                  hand_left_data[frame][0][2] -pose_data[frame][15][2]]
        r_disp = [hand_right_data[frame][0][0] - pose_data[frame][16][0],
                  hand_right_data[frame][0][1] - pose_data[frame][16][1],
                  hand_right_data[frame][0][2] - pose_data[frame][16][2]]

        
        for joint in range(len(hand_left_data[frame])):
            hand_left_data[frame][joint] = (hand_left_data[frame][joint][0] - l_disp[0],hand_left_data[frame][joint][1] - l_disp[1],hand_left_data[frame][joint][2] - l_disp[2])

            hand_right_data[frame][joint] = (hand_right_data[frame][joint][0] - r_disp[0], hand_right_data[frame][joint][1] - r_disp[1], hand_right_data[frame][joint][2] - r_disp[2])

                    
    return pose_data,hand_left_data,hand_right_data

def mediapipe_to_posenet(pose):
    list_poses_to_mantain = [0,2,5,7,8,11,12,13,14,15,16,23,24]
    posenet = []
    for frame in pose:
        framenet = []
        for n,bone in enumerate(frame):
            if n in list_poses_to_mantain:
                framenet.append(bone)
        framenet.append(frame[23])
        framenet.append(frame[24])
        framenet.append(frame[23])
        framenet.append(frame[24])
        posenet.append(framenet)
    return posenet

def chicken_neck_mediapipe(pose,lhand,rhand):
    for n,frame in enumerate(pose):
        disp = np.array([0,0,0]) - (np.array(frame[5]) + np.array(frame[6]))*0.5
        for i in range(len(pose[n])):
            pose[n][i] = list(np.array(pose[n][i]) + disp)
            pose[n][i][1] = pose[n][i][1] - 1
        for i in range(len(lhand[n])):
            lhand[n][i] = list(np.array(lhand[n][i]) + disp)
            lhand[n][i][1] = lhand[n][i][1] - 1
        for i in range(len(rhand[n])):
            rhand[n][i] = list(np.array(rhand[n][i]) + disp)
            rhand[n][i][1] = rhand[n][i][1] - 1
    return pose, lhand,rhand

def mediapipe_to_posenet_dict(pose,lhand,rhand):
    dict_posenet = deepcopy(output_json)
    for frame in pose:
        for n,bone in enumerate(frame):
            dict_posenet['posenet'][pose_config[n]]['x'].append(bone[0])
            dict_posenet['posenet'][pose_config[n]]['y'].append(bone[1])
            dict_posenet['posenet'][pose_config[n]]['z'].append(bone[2])
    for frame in lhand:
        for n,bone in enumerate(frame):
            dict_posenet['leftHand'][hand_config[n]]['x'].append(bone[0])
            dict_posenet['leftHand'][hand_config[n]]['y'].append(bone[1])
            dict_posenet['leftHand'][hand_config[n]]['z'].append(bone[2])
    for frame in rhand:
        for n,bone in enumerate(frame):
            dict_posenet['rightHand'][hand_config[n]]['x'].append(bone[0])
            dict_posenet['rightHand'][hand_config[n]]['y'].append(bone[1])
            dict_posenet['rightHand'][hand_config[n]]['z'].append(bone[2])
    return dict_posenet
```

```
import math
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
from datetime import datetime
from unicodedata import normalize
import tempfile
import re
import json
import hashlib
import os
import copy
from typing import Dict, List, Tuple
import time


def extract_keypoints_func(filepath: str):
    pose_data,hand_left_data,hand_right_data,face_data = extract_keypoints(filepath)
    pose_keypoints_data, hand_left_keypoints_data, hand_right_keypoints_data = interpolate_keypoints([pose_data,hand_left_data,hand_right_data])
    pose_keypoints_data, hand_left_keypoints_data, hand_right_keypoints_data = put_hand_in_body(pose_keypoints_data,hand_left_keypoints_data,hand_right_keypoints_data)
    pose_keypoints_data = mediapipe_to_posenet(pose_keypoints_data)
    pose_keypoints_data, hand_left_keypoints_data, hand_right_keypoints_data = chicken_neck_mediapipe(pose_keypoints_data, hand_left_keypoints_data, hand_right_keypoints_data)
    data = mediapipe_to_posenet_dict(pose_keypoints_data, hand_left_keypoints_data, hand_right_keypoints_data)
    cj.FileIO.create(f'{filepath.split(".")[0]}.json', data).save(exist_ok=True)


# realiza download de uma arquivo de vídeos do storage
def download_video(video_id: str, video_filename: str, dest_path: str, workspace: str, language_id: str):
    bucket_path = f'{workspace}/{language_id}/videos/{video_id}/{video_filename}.mp4'
    Community.firebase.storage.db.blob(bucket_path).download_to_filename(dest_path)

def download_json(video_id: str, dest_path: str, folder_lang: str):
    bucket_path = f'modelos/mediaPipe/{folder_lang}/{video_id}.json'
    Community.firebase.storage.db.blob(bucket_path).download_to_filename(dest_path)

# realiza upload para o firebase
def upload_json(video_id: str, filepath: str, workspace: str, language_id: str):
    blob = Community.firebase.storage.db.blob(f'{workspace}/{language_id}/keypointsJson/{video_id}.json')
    blob.upload_from_filename(filepath)

# converte frames e cropa o video
def convert_and_crop_video(filepath: str):
    import os
    out_temp_path = tempfile.NamedTemporaryFile(suffix='.mp4')
    os.system(f"ffmpeg -i '{filepath}' -f webm -r 30 -qscale 0 -vf crop='if(gte(ih\,iw)\,in_w\,in_h)':in_h,pad=ih:ih:'if(gte(ih\,iw)\,((ih-iw)/2)*-1\,0)':0,scale=1080:1080 -f mp4 '{out_temp_path.name}' -hide_banner -loglevel error -y")
    return out_temp_path
```

```
from cereja import Progress
from IPython.display import clear_output

for index in Progress.prog(range(next_index,len(videos_to_extract))):
    video_id = videos_to_extract[index]
    video_ref = cmt.get_document('videos',video_id)
    video_file = video_ref.midia[0]['fileName']

    # cria path temporário
    temp_path = tempfile.NamedTemporaryFile(suffix='.mp4')
  
    downloaded = True

    # baixar video
    try:
        download_video(video_id, video_file, temp_path.name, WORKSPACE, LANGUAGE_ID)
    except:
        print(f'\tDownload error!:\n\tverifique o vídeo em\n\t/{WORKSPACE}/{LANGUAGE_ID}/videos/{video_id}')
        downloaded = False

    if downloaded:
        
        # converter e subir vídeo
        converted_temp_path = convert_and_crop_video(temp_path.name)

        # extrair kp
        extract_keypoints_func(converted_temp_path.name)

        # fazer upload
        json_path = f'{converted_temp_path.name.split(".")[0]}.json'
        
        if os.path.exists(json_path):
            upload_json(video_id, json_path, WORKSPACE, LANGUAGE_ID)
            update = [(video_ref.reference,{'keypointsExtracted': True})]
            cmt.firestore.update(update,safe_mode=False)
            # exclui arquivo temporário
            !rm '{json_path}'
        else:
            print(f'\tErro de extração:\n\tverifique o vídeo em\n\t/{WORKSPACE}/{LANGUAGE_ID}/videos/{video_id}')

        # exclui arquivos temporários
        converted_temp_path.close()
        temp_path.close()
        cj.FileIO.create(f'/content/drive/Shareddrives/IA/CIENCIA_DE_DADOS/PROCESSOS/EXTRACT_KP/progress_{user}.txt',[index]).save(exist_ok=True)
    clear_output(wait = True)
```
